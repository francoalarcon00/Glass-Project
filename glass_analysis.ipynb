```python
description_RI = lv.pd.DataFrame(df.RI.describe())
print('')
print(description_RI.T)
print('')
lv.sns.set_theme(style='darkgrid')
lv.plt.figure(figsize=(20, 15))
lv.plt.subplot(2, 1, 1)
lv.sns.histplot(x='RI', kde=True, data=df, bins=20, color='indianred')
lv.plt.subplot(2, 1, 2)
lv.sns.boxplot(x='RI', data=df, color='indianred')
```
