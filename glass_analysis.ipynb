```python

lv.plt.figure(figsize=(18, 24))
lv.sns.set_theme(style='darkgrid')

cols = ['RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe']
i=0

# Muestro las distribuciones de cada columna
for col in cols:
    colors = ['indianred', 'yellowgreen', 'dodgerblue', 'goldenrod', 'darkslategray', 
                'purple', 'mediumseagreen', 'darkorange', 'hotpink']
    lv.plt.subplot(5, 2, cols.index(col) + 1)
    lv.sns.histplot(x=col, kde=True, data=df, bins=20, color=colors[i])
    i+=1
```

```python
lv.plt.figure(figsize=(12,5))
lv.sns.countplot(x='Type', data=df, palette='Spectral')
```

```python
lv.plt.figure(figsize=(18,25))
for col in cols:
    lv.plt.subplot(5, 2, cols.index(col)+1)
    lv.sns.histplot(data=df, x=col, element='poly', hue='Type', bins=6, palette='Spectral')
```
