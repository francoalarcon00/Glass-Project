<h3 align='center'>RI: Indice de refracci√≥n üëà</h3>

```python
description_RI = pd.DataFrame(df.RI.describe())
print('')
print(description_RI.T)
print('')
sns.set_theme(style='darkgrid')
plt.figure(figsize=(20, 10))
plt.subplot(2, 1, 1)
sns.histplot(x='RI', kde=True, data=df, bins=20, color='indianred')
plt.subplot(2, 1, 2)
sns.boxplot(x='RI', data=df, color='indianred')
plt.show()
```

<h4 align="center">Observaciones üïµÔ∏è‚Äã </h4>
<p>
    <li>La media del RI es de 1.518 unidades aproximadamente (el RI no posee unidad f√≠sica).</li>
    <br>
    <li>La distribuci√≥n posee un bajo desv√≠o est√°ndar (0.003 aprox), es decir, no existe una variaci√≥n significativa de los datos. Sin embargo, se puede visualizar una tendencia sesgada levemente hacia la derecha tanto en el boxplot como en el histograma.</li>
    <br>
    <li>La mediana se encuentra cerca de los 0.152 unidades.</li>
    <br>
    <li>Observemos que existen varios outliers en esta variable.</li>
</p>

<h3 align='center'>Types: Tipos üëà</h3>

```python
plt.figure(figsize=(20,10))
sns.countplot(x='Type', data=df, palette='Spectral')
plt.show()
```

<h4 align="center">Observaciones üïµÔ∏è‚Äã </h4>
<p>
    <li>Vemos una predominancia casi total de los tipos 1 y 2</li>
    <li>No se registraron datos de tipo 4</li>
    <li>El tipo 6 es el que menos se registr√≥</li>
</p>

```python
plt.figure(figsize=(22,40))

cols = ['RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe']

for col in cols:
    plt.subplot(5, 2, cols.index(col)+1)
    sns.histplot(data=df, x=col, element='poly', hue='Type', bins=6, palette='Spectral')
    plt.title('Feature ('+str(col)+') por tipos', fontdict={'fontsize': 18})
plt.show()
```

<h4 align="center">Observaciones üïµÔ∏è‚Äã </h4>
<p>
    <li>Como explicamos anteriormente, vemos esa predominancia casi total en todos los features de los tipos 1 y 2</li>
    <br>
</p>
